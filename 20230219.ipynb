## 複習一下 requests 的使用

從政府資料開放平台抓資料：https://data.gov.tw/

### 練習
抓取年度國內金融指標做分析:https://apiservice.mol.gov.tw/OdService/download/A17000000J-030244-dpp

````python
import requests
````

````python
url = "https://apiservice.mol.gov.tw/OdService/download/A17000000J-030244-dpp"
````

````python
data = requests.get(url).json()
````

````python
data
````

````python
#檢視一下data的資料型態
type(data)
````

````python
#檢視一下data裡面有幾筆資料
len(data)
````

````python
# 檢視一下 data 的第一筆資料
data[0]
````

## 列出各年度的M2貨幣供給額

### 方法一:

````python
for i in data:
  print(i['年度'], "==>", i['貨幣供給額期底年增率-M2'])
````

### 方法二:

````python
for i in range(len(data)):
  print(i)
````

````python
for i in range(len(data)):
  print(data[i]['年度'], "==>", data[i]['貨幣供給額期底年增率-M2'])
````

````python
for i in range(len(data)):
  print(data[i]['年度'], "==>", data[i]['證券（集中市場）-發行量加權股價指數（月底）'])
````
